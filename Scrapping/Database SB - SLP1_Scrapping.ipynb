```python
from bs4 import BeautifulSoup as soup
import requests
import re
import pandas as pd
from collections import defaultdict
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate
  

```

```python
r = requests.get('https://vedabase.io/en/library/sb/')
main= soup(r.text,'lxml')
mstr1= 'https://vedabase.io/en/library/sb/'
mstr = 'https://vedabase.io'
sbdict =  defaultdict(list)
```

```python
cont_list =main.find_all('div', class_="col-6 col-sm-4 col-md-3 col-lg-2 col-lg-15 text-center book-item")
print(cont_list)
```

```python
s_cont = str(cont_list)
```

```python
f='https://vedabase.io/en/library/sb/1/1/1'

while(re.search('/en/library/sb/.{0,2}/',str(cont_list))):
    #Search for contents in main library
    x_cont=re.search('/en/library/sb/.{0,2}/',str(cont_list))
    #found a match of chapter 
    p1=str(x_cont.group())
    
    
    
    print(p1)
    while(re.search(p1,str(cont_list))):
        print(re.search(p1,str(cont_list)))
        cont_list= re.sub(p1,'',str(cont_list),1)
    url1 = mstr + p1
    print(url1)
    r1 = requests.get(url1)
    deep1= soup(r1.text,'lxml')
    #Entered into Canto
    chap_list =deep1.find_all('div', class_="r-chapter")
    rand2 = p1+'.{0,2}/'
    while(re.search(rand2,str(chap_list))):
        x_chap=re.search(rand2,str(chap_list))
        
        p2= str(x_chap.group())
        chap_list= re.sub(p2,'',str(chap_list),1)
        print(p2)
        url2 = mstr+p2
        if(f==url2):
            continue
        f=url2
        
        print(url2)
        r2= requests.get(url2)
        deep2=soup(r2.text,'lxml')
        #entered a chapter
        ver_list=deep2.find_all('dt')
        rand_var = p2+'.*?/'
        print(rand_var)
        
        while(re.search(rand_var,str(ver_list))):
            x1=re.search(rand_var,str(ver_list))
            p2=str(x1.group())
            ver_list = re.sub(p2,'',str(ver_list),1)
            url3= mstr + p2
             
           
            print(url3)
            #entered a verse
            r3=requests.get(url3)
            deep3 = soup(r3.text,'lxml')
            
            #Column 2(shloka)
            
            shloka= deep3.find_all('div', class_="wrapper-verse-text")
            while(re.search('<.+?>',str(shloka))):
                shloka = re.sub('<.+?>','',str(shloka),1)
            shloka = re.sub('Text','',str(shloka),1)
            sbdict[(re.search('sb/.{0,2}/.*?/.*?/',p2)).group()].append(str(shloka))
            
            #column 3(shloka SLP)
            s1 = transliterate(str(shloka), sanscript.IAST, sanscript.SLP1)
            sbdict[(re.search('sb/.{0,2}/.*?/.*?/',p2)).group()].append(str(s1))
            #column 4 (shloka without space)
            sbdict[(re.search('sb/.{0,2}/.*?/.*?/',p2)).group()].append(str(shloka).replace(" ", ""))
            #column 5(translation)
            transl= deep3.find_all('div', class_="wrapper-translation")
            
            while(re.search('<.+?>',str(transl))):
                transl = re.sub('<.+?>','',str(transl),1)
            transl = re.sub('Translation','',str(transl),1)
            sbdict[(re.search('sb/.{0,2}/.*?/.*?/',p2)).group()].append(str(transl))
            #column 6(puport with tags SLP)
        
        
            purport= deep3.find_all('div', class_="wrapper-puport")
            purport = re.sub('Purport','',str(purport),1)
            s4 = transliterate(str(purport), sanscript.IAST, sanscript.SLP1)
            sbdict[(re.search('sb/.{0,2}/.*?/.*?/',p2)).group()].append(str(s4))
            #column 7 (purport without tags slp)
            while(re.search('<.+?>',str(s4))):
                s4 = re.sub('<.+?>','',str(s4),1)
            
            sbdict[(re.search('sb/.{0,2}/.*?/.*?/',p2)).group()].append(str(s4))
            
            #column 8 (purport normal with tags)
            sbdict[(re.search('sb/.{0,2}/.*?/.*?/',p2)).group()].append(str(purport))
            
            
            #Column 9 (extracts)
            ext= deep3.find_all('em')
            s5 = transliterate(str(ext), sanscript.IAST, sanscript.SLP1)
            sbdict[(re.search('sb/.{0,2}/.*?/.*?/',p2)).group()].append(str(s5))
            
            #Column 10 (Patterns)
            
            
        
        
        
        
        
        
            
            
    
    
    
    
    


    
```

```python
df=pd.DataFrame.from_dict(sbdict,orient='index')
```
